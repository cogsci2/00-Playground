```python
df.info()
```

```python
df = df.query("written_review != '' ")  # only with written reviews
df = df.query('guide_status == "Local Guide"') # only local guides
df.dropna(subset=['written_review'], inplace=True) # drop missing values
df.sort_values(by=['num_reviews'], inplace=True, ascending=False) # sort by num_reviews
df.reset_index(drop=True, inplace=True) # reset index
df
```
